## Basic settings

```python
import os, sys, time
from IPython.display import display
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path: sys.path.append(module_path)
from src.utl.utl import UTC
from src.utl.read_files import check_dir
from src.utl.widgets import html_table
from src.utl.pqa_table import table_cds
tmpath = module_path + '/tmp/'
if check_dir(tmpath, True, 'none'):
    table = []
    long_table = []
    header = ['Line','Corpus','Parsing','LW','"."','Generalization','Space','Rules','PA','PQ']
    start = time.time()
    print(UTC(), ':: module_path =', module_path)
else: print(UTC(), ':: could not create temporary files directory', tmpath)
```

## Corpus test settings

```python
out_dir = module_path + '/output/Child-Directed-Speech-' + str(UTC())[:10]
runs = (1,1)    # (attempts to learn grammar per line, grammar tests per attempt)
if runs != (1,1): out_dir += '-multi'
# cp,rp: corpus_path, reference_path:
rp = module_path + '/data/ChildDirectedSpeech/LG-English'
cp = rp  # corpus path = reference_path :: use 'gold' parses as test corpus
kwargs = {
    'left_wall'     :   ''          ,
    'period'        :   False       ,
    'clustering'    :   'kmeans'    ,
    'cluster_range' :   (30,121,2)  ,  # (4,51,2) updated 2018-07-20
    'cluster_criteria': 'silhouette',
    'cluster_level' :   1           ,  # 1: use clusters number with max silhouette
    'tmpath'        :   tmpath      , 
    'verbose'       :   'min'       ,
    'template_path' :   'poc-turtle',
    'linkage_limit' :   1000         ,
    'categories_generalization': 'off' }
lines = [
    [49, 'ChildDirectedSpeech' , 'LG-English'                     ,0,0, 'none'  ], 
    [50, 'ChildDirectedSpeech' , 'LG-English'                     ,0,0, 'rules' ], 
    [51, 'ChildDirectedSpeech' , 'R=6-Weight=6:R-mst-weight=+1:R' ,0,0, 'none'  ], 
    [52, 'ChildDirectedSpeech' , 'R=6-Weight=6:R-mst-weight=+1:R' ,0,0, 'rules' ]]
```
